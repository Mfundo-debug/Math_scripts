::: callout-note

 plot a sine function

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
#plot cos function using the above data
y = np.cos(x)
plt.plot(x, y)
plt.show()  # show the plot on the screen
```

```python
#integrate sinx and cosx using sympy
from sympy import *
x = symbols('x')
integrate(sin(x), x)
integrate(cos(x), x)
# show the equation in display math format
from IPython.display import display, Math, Latex
display(Math(r'$$\int sin(x)dx = -cos(x) + C$$'))
display(Math(r'$$\int cos(x)dx = sin(x) + C$$'))
display(Math(r'$$\frac{d}{dx}sin(x) = cos(x)$$'))
display(Math(r'$$\frac{d}{dx}cos(x) = -sin(x)$$'))
```
